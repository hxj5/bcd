```python
# mtx2adata.ipynb - convert sparse matrix file into adata file.
```

```python
import os
import pandas as pd
from sccnasim.xlib.xdata import load_10x_data
```

```python
data_dir = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/sccnasim-cs_screadsim/rdr_starsolo"
gene_anno_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/base/data/gene_anno/refdata-cellranger-GRCh38-3.0.0.valid_name.uniq_name.chrom1-22XY.5column.genes.tsv"
cell_anno_fn = '/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/sccnasim-cs_screadsim/simu/final/spot_anno.tsv'

out_dir = data_dir
out_adata_fn = os.path.join(out_dir, "HCC3N_simu.spotxgene.starsolo.h5ad")
```

## Construct adata

```python
adata = load_10x_data(data_dir)
adata
```

```python
cell_anno = pd.read_csv(cell_anno_fn, header = None, sep = '\t')
cell_anno.columns = ['cell', 'cell_type']
cell_anno
```

```python
cell_anno['cell_type'].value_counts()
```

```python
adata.obs = adata.obs.merge(cell_anno, on = 'cell', how = 'left')
adata.obs
```

```python
adata.obs['cell_type'].value_counts()
```

```python
adata.var.columns = ['gene_id', 'gene']
adata.var
```

```python
gene_anno = pd.read_csv(gene_anno_fn, sep = '\t', header = None, dtype = {0:str})
gene_anno.columns = ['chrom', 'start', 'end', 'gene', 'strand']
gene_anno
```

```python
adata.var = adata.var.merge(gene_anno, on = 'gene', how = 'left')
adata.var
```

## Save adata

```python
adata.X = adata.X.toarray()
```

```python
adata
```

```python
adata.write(out_adata_fn, compression = 'gzip')
```

```python
adata.obs.to_csv(
    os.path.join(out_dir, 'spot_anno.tsv'),
    header = False,
    index = False,
    sep = '\t'
)
```
