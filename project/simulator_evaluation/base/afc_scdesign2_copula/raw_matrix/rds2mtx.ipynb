```r
# rds2mtx.ipynb - convert .rds file into .mtx file.
```

```r
library(Matrix)
```

```r
simu_count_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/afc_scdesign2_copula/simu/scDesign2_copula.simu_count.rds"
out_matrix_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/afc_scdesign2_copula/raw_matrix/matrix.mtx"
```

```r
seed_count_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/seed/afc/HCC3N_600spot.spotxgene.afc.rds"
gene_anno_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/base/data/matrix/genes.tsv"
out_gene_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/afc_scdesign2_copula/raw_matrix/genes.tsv"
```

## Save to `.mtx` file

```r
dat = readRDS(simu_count_fn)
str(dat)
```

```r
dat = as(dat, "sparseMatrix")
str(dat)
```

```r
writeMM(dat, out_matrix_fn)
```

## Re-generate the gene annotations

```r
seed = readRDS(seed_count_fn)
str(seed)
```

```r
nonspikes = which(!grepl("ercc", colnames(seed), ignore.case = TRUE))
dat_genes = colnames(seed)[nonspikes]
str(dat_genes)
```

```r
gene_anno = read.table(gene_anno_fn, header = FALSE, stringsAsFactors = FALSE)
colnames(gene_anno) = c("gene_id", "gene_name")
str(gene_anno)
```

```r
rownames(gene_anno) = gene_anno$gene_name
anno_s = gene_anno[dat_genes, ]
str(anno_s)
```

```r
write.table(anno_s, out_gene_fn, sep = '\t', quote = FALSE, row.names = FALSE, col.names = FALSE)
```
