```python
import anndata as ad
import numpy as np
import os
import pandas as pd
import scanpy as sc
```

```python
spaceranger_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/seed/spaceranger/HCC3N_600spot.spotxgene.h5ad"
starsolo_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/seed/rdr_starsolo/HCC3N_600spot.spotxgene.starsolo.h5ad"
afc_fn = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/base/gen_data/sccnasim/simu/2_afc/afc.counts.cell_anno.h5ad"

out_dir = "/groups/cgsd/xianjie/projects/cna-benchmark/HCC3N_600spot/simulator_evaluation/internal_consistency/seed_tools/pp"
```

```python
normal_cell_type = "normal"
```

# Load Data

```python
os.makedirs(out_dir, exist_ok = True)
```

### SpaceRanger

```python
spaceranger = ad.read_h5ad(spaceranger_fn)
spaceranger
```

```python
spaceranger.obs
```

```python
spaceranger.var
```

```python
spaceranger = spaceranger[:, ~spaceranger.var["feature"].duplicated(keep = False)].copy()
spaceranger.obs.index = spaceranger.obs["cell"]
spaceranger.var.index = spaceranger.var["feature"]
spaceranger
```

### STARsolo

```python
starsolo = ad.read_h5ad(starsolo_fn)
starsolo
```

```python
starsolo.var['feature'] = starsolo.var['gene']
starsolo = starsolo[:, ~starsolo.var["feature"].duplicated(keep = False)].copy()
starsolo.obs.index = starsolo.obs["cell"]
starsolo.var.index = starsolo.var["feature"]
starsolo
```

### scCNASim afc

```python
afc = ad.read_h5ad(afc_fn)
afc
```

```python
afc = afc[:, ~afc.var["feature"].duplicated(keep = False)].copy()
afc.X = afc.layers["A"] + afc.layers["B"] + afc.layers["U"]
afc = afc[afc.obs["cell_type"] == normal_cell_type, :].copy()
afc.obs.index = afc.obs["cell"]
afc.var.index = afc.var["feature"]
afc
```

### Use the same order of cells

```python
spaceranger = spaceranger[afc.obs.index, :].copy()
spaceranger
```

```python
starsolo = starsolo[afc.obs.index, :].copy()
starsolo
```

## Use intersect genes

```python
adata_list = [spaceranger, starsolo, afc]

genes = None
for i, adata in enumerate(adata_list):
    if i == 0:
        genes = adata.var["feature"].to_numpy()
    else:
        genes = np.intersect1d(genes, adata.var["feature"])
genes.shape
```

### Use the same order of genes

```python
afc = afc[:, afc.var["feature"].isin(genes)].copy()
print(afc)

spaceranger = spaceranger[:, afc.var.index].copy()
print(spaceranger)

starsolo = starsolo[:, afc.var.index].copy()
print(starsolo)
```

# Save Data

### Check the order of cells and genes

```python
assert np.all(spaceranger.obs["cell"].to_numpy() == afc.obs["cell"].to_numpy())
assert np.all(starsolo.obs["cell"].to_numpy() == afc.obs["cell"].to_numpy())
```

```python
assert np.all(spaceranger.var["feature"].to_numpy() == afc.var["feature"].to_numpy())
assert np.all(starsolo.var["feature"].to_numpy() == afc.var["feature"].to_numpy())
```

```python
spaceranger.var.index.name = None
starsolo.var.index.name = None
afc.var.index.name = None
```

```python
spaceranger.write_h5ad(os.path.join(out_dir, "spaceranger.h5ad"), compression = 'gzip')
starsolo.write_h5ad(os.path.join(out_dir, "starsolo.h5ad"), compression = 'gzip')
afc.write_h5ad(os.path.join(out_dir, "afc.h5ad"), compression = 'gzip')
```
